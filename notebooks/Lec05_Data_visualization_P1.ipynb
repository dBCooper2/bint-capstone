## $\color{red}{\text{Data Visualization}}$
1. **Data visualization** is the use of charts to form a basic understanding of data
2. It is an effective part of any data analysis
3. Often, the data type will dictate which chart to use
    - Example: Numeric, Categorical, etc.

## $\color{red}{\text{Categories of Data Visualization}}$
### $\color{blue}{\text{Univariate and Multivariate Charts}}$

1. **Univariate charts**
    - Displays the behavior of a single variable
2. **Multivariate charts**
    - Displays the relationship between two or more variables

## $\color{red}{\text{Univariate Charts For Categorical Data}}$
### $\color{blue}{\text{Displaying Counts: Bar Plot}}$

1. For categorical data, there are limited options for visualization
2. Counts and percentages to display values in categories

### $\color{green}{\text{Example 1}}$

1. Provide a bar graph that answers the following:
2. Comment on it
    - What are the proportions of males and females in the HR dataset?

```python
# value_counts() returns a table of the values in each column
# Normalize=True gives a pct
x = df['Gender'].value_counts(normalize=True) 
x
```

```python
count_m = df['Gender'].value_counts()['Male']
count_f = df['Gender'].value_counts()['Female']
count_m, count_f
```

```python


# Add a Title
plt.title('Percentage of Gender')
plt.ylabel('Percentage (0-1 Scale)')

# Color Sequences: Add Colors to the Chart (from L to R)
my_colors = list(['orange','blue']) # colors as a list
my_colors = list(['orange','#2142ab']) # colors with hex
my_colors = list('mc') # colors codes (magenta, cyan)

# instantiate a bar chart of our normalized gender data
x.plot(kind='bar', color=my_colors) 
plt.show() # shows only the plot
```

## $\color{red}{\text{Univariate Charts For Numeric Data}}$
### $\color{blue}{\text{Histogram}}$

1. A histogram is useful for showing the distribution of data
2. It is possible to plot more than one (1) histogram on a chart

### $\color{green}{\text{Example 3}}$
1. Provide a histogram of monthtly income and commment on it
2. Provide a histogram of monthly rate and comment on it
    - Does anything stand out?

```python
x = df['MonthlyIncome']

x.plot(kind='hist', bins=50)
```

```python
# Alternatively...

# bins sets the bandwidth, alpha sets the opacity of the chart
x.hist(bins=50, alpha=.8)
plt.xlabel('Income')
plt.ylabel('# of Employees')
plt.title('Monthly Income :: Probability Distribution', fontsize=12)
plt.show()
```

Plotting Multiple Hists at Once:

```python
x1 = df['MonthlyIncome']
x2 = df['MonthlyRate']

x1.plot(kind='hist', bins=50, alpha=.5)
x2.plot(kind='hist', bins=50, alpha=.5)

plt.legend(loc='upper right')
plt.xlabel('Income', fontsize=12)
plt.title('Histogram of Monthly Rate and Income')
plt.show()
```

**In this example, you need to be mindful of the scale! This is a bad chart, the variables you are plotting together need to make sense!**

Also, look at how it automatically chose Orange/Blue, that's pretty neat.

### $\color{blue}{\text{Density Plot}}$
1. Gives an estimation of probability density function
2. The **area** under the curve must sum to 1
3. The probability of a number between two values is the area under the curve of the two points

A density function is any function that returns a probability.
