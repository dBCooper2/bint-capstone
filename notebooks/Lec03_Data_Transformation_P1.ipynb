## Filtering by Columns

```python
# Creating a Subset using Column Names
cols = ['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department']

sub_df = df[cols]
sub_df.head()
```

```python
# Creating a Subset using Column Numbers

sub_df = df.iloc[:,0:6] # iloc = integer location. : grabs every row, 0:6 grabs the first 5 columns
sub_df.head()
```

## Filter by Rows

```python
# You can Filter by Rows using iloc as well
sub_df = df.iloc[0:6, :] # This time we grab only the first 5 rows, but grab every column
sub_df
```

A note on `iloc[]`: when you want data in a range of `n:m`, make sure you are specifying the range as `n:m+1`. The first index is inclusive, but the last index is exclusive; i.e. if I want the first 2 rows I need to access 0:2. This pulls row 0 and row 1, and then hits row 2 and exits without adding row 2 to the grabbed data

## $\color{red}{\text{Grouping and Aggregating Data}}$
### $\color{blue}{\text{SAC Method: Split, Apply and Combine}}$

SAC.PNG

#### $\color{green}{\text{Example 5}}$
1. Apply the SAC method to find the following:
    - Average **Age** by **Attrition**
    - Median **Age** by **Attrition**
    - Minimum **Age** by **Attrition**
    - Maximum **Age** by **Attrition**
2. Comment on it
    - Is there any evidence that age is a factor among those who leave the company?

```python
vars_of_interest = ['Age', 'Attrition']
df_sac = df[vars_of_interest]
df_sac.head()
```

## Average Age by Attrition

```python
grouped = df_sac.groupby('Attrition') # SPLIT (or group by a categorical variable)
avg_age = grouped.mean() # APPLY: perform operations on the groups
avg_age # return as a COMBINED df of the groups
```

## Median Age by Attrition

```python
grouped = df_sac.groupby('Attrition') # SPLIT (or group by a categorical variable)
med_age = grouped.median() # APPLY: perform operations on the groups
med_age # return as a COMBINED df of the groups
```

## Minimum Age by Attrition

```python
grouped = df_sac.groupby('Attrition') # SPLIT (or group by a categorical variable)
min_age = grouped.min() # APPLY: perform operations on the groups
min_age # return as a COMBINED df of the groups
```

## Maximum Age by Attrition

```python
grouped = df_sac.groupby('Attrition') # SPLIT (or group by a categorical variable)
max_age = grouped.max() # APPLY: perform operations on the groups
max_age # return as a COMBINED df of the groups
```
