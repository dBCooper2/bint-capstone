```python
import numpy as np
import pandas as pd

# Number of minutes you want to simulate
minutes = 48

# Random realistic simulation ranges for each stat
data = {
    'field_goals_attempted': np.random.randint(0, 4, size=minutes),   # 0-3 FGA per minute
    'three_pointers_attempted': np.random.randint(0, 2, size=minutes), # 0-1 3PA per minute
    'free_throws_attempted': np.random.randint(0, 2, size=minutes),    # 0-1 FTA per minute
    'rebounds_offensive': np.random.randint(0, 2, size=minutes),       # 0-1 OREB per minute
    'rebounds_defensive': np.random.randint(0, 3, size=minutes),       # 0-2 DREB per minute
    'steals': np.random.binomial(1, 0.05, size=minutes),               # ~5% chance per minute
    'blocks': np.random.binomial(1, 0.04, size=minutes),               # ~4% chance per minute
    'turnovers': np.random.binomial(1, 0.08, size=minutes),            # ~8% chance per minute
    'fouls_personal': np.random.binomial(1, 0.06, size=minutes),        # ~6% chance per minute
    'estimated_pace': np.random.normal(loc=100, scale=5, size=minutes), # average pace ~100
    'pace': np.random.normal(loc=100, scale=5, size=minutes),           # pace again
    'distance': np.random.normal(loc=5, scale=1, size=minutes),         # average 5 miles covered
    'rebound_chances_offensive': np.random.randint(0, 3, size=minutes), # 0-2 OREB chances
    'secondary_assists': np.random.binomial(1, 0.1, size=minutes),      # ~10% chance per minute
    'free_throw_assists': np.random.binomial(1, 0.05, size=minutes),    # ~5% chance per minute
}

# Create the DataFrame
simulated_game_df = pd.DataFrame(data)

# Make sure no negative values (in case of weird normal distributions)
simulated_game_df = simulated_game_df.clip(lower=0)

simulated_game_df.head()
```
