# NBA Shots :: Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_pickle('/Users/dB/Documents/repos/github/bint-capstone/data-sources/nba/nba_player_stats_2004_2024.pkl')
```

```python
# create correlation matrix
matr = df.select_dtypes(include=['number']).corr()


# plot
plt.figure(figsize=(8,6))
sns.heatmap(matr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
